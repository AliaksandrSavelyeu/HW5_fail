```python
df = data.drop(columns = ['id', 'url', 'region_url', 'VIN', 'size', 'image_url', 'description', 'county', 'lat', 'long', 'posting_date', 'title_status'])
```

```python
df.isnull().sum()
```

```python
mode_fare = df['condition'].mode()[0]
df['condition'].fillna(mode_fare, inplace=True)

mode_fare = df['cylinders'].mode()[0]
df['cylinders'].fillna(mode_fare, inplace=True)

mode_fare = df['drive'].mode()[0]
df['drive'].fillna(mode_fare, inplace=True)

mode_fare = df['paint_color'].mode()[0]
df['paint_color'].fillna(mode_fare, inplace=True)

mode_fare = df['type'].mode()[0]
df['type'].fillna(mode_fare, inplace=True)
```

```python
df
```

```python
df_prep = df.dropna()
```

```python
df_prep.isnull().sum()
```

```python
df_prep.columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
df_prep[['region', 'manufacturer', 'model', 'condition',
    'cylinders', 'fuel', 'transmission','drive',
    'type', 'paint_color', 'state']] = encoder.fit_transform(df_prep[['region', 'manufacturer', 'model', 'condition',
    'cylinders', 'fuel', 'transmission','drive',
    'type', 'paint_color', 'state']])
```
